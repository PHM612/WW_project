## Features


    - Retrieves stream data from the USGS NWIS service
    - Retrieve data using multiple site numbers, state, county codes, or boundary box
    - Preserves NWIS metadata, including NoData values
    - Helpful error messages to help you write valid requests
    - Extracts data into a Pandas dataframe, json, or dict
    - Plotting and manipulation through Pandas dataframes
    - Interactive map for finding stream gage ID numbers
    - Still in early development! More features to come!

```python
# request data from the USGS National Water Information System (NWIS):
# site = '01582500'
# service = 'dv'
# start_date = '2015-05-10'
# end_date = '2015-05-15'
# stateCd = None
# countyCd = None
# bBox = None
# parameterCd = '00060'
# period = None

# Exemple 1
D1 = ['01585200', 'dv', None, None, None, None, None, '00060', 'P55D']

# Exemple 2
D2 = ['01582500', 'dv', '2015-05-10', '2015-05-15', None, None, None, '00060', None]

# Exemple 3
# Using the NWIS object to request data
# use the NWIS object to store your response and the extracted data.
D3 = ['01585200', 'dv', None, None, None, None, None, '00060', 'P20D']
# set up our system, now we submit our request for data:
```

```python
# Create NWIS data object to hold our reques
data = hf.NWIS(D1[0],D1[1],D1[2],D1[3],D1[4],D1[5],D1[6],D1[7],D1[8])
# data = hf.NWIS(site, service, start_date, end_date,
#    stateCd, countyCd, bBox, parameterCd, period)
# We've set up our system, now we submit our request for data and get a response
data.get_data()
# answer should be <hydrofunctions.station.NWIS at 0x127506d6ac8>
data.df().head()
# Create a dataframe from our data, and list the first five items
```

## Autre façon d'appeler les données

```python
# autre façon d'appeler les données
# format json - Examine the response
# Extract the full json response from the data provider
data1 = hf.get_nwis(D2[0],D2[1],D2[2],D2[3],D2[4],D2[5],D2[6],D2[7],D2[8])
print(data1.ok)
print(data1.status_code)
print()
# JavaScript Object Notation (json)
# Format de données textuelles dérivé de la notation des objets du langage JavaScript.
# Il permet de représenter de l’information structurée comme le permet XML par exemple. 
# Actuellement décrit par 2 normes concurrentes RFC 8259 [archive] de l’IETF et ECMA-4041 de l'ECMA
# Des bibliothèques pour le format JSON existent dans la plupart des langages de programmation.
my_dict = data1.json()
print(my_dict)
print()
print(data1.headers)
```

```python
# list of all attributes with hf.get_nwis type
dir(data1)
```

## Multiple sites, other parameters

```python
# It's possible to load data from several different sites 
# at the same time, and you aren't limited to just stream discharge.
# Request data for several sites in a list
sites = ['380616075380701','394008077005601']
# The NWIS can deliver data as daily mean values ('dv') or as instantaneous values ('iv') 
# that can get collected as often as every five minutes!
service = 'iv'
# Depending on the site, the USGS collects groundwater levels ('72019'), stage ('00065'), precipitation, and more!
pcode = '72019'
# create a new dataset called 'groundwater', and we'll use the values we set up above.

# request data from the USGS National Water Information System (NWIS):
# site = '01582500'
# service = 'dv'
# start_date = '2015-05-10'
# end_date = '2015-05-15'
# stateCd = None
# countyCd = None
# bBox = None
# parameterCd = '00060'
# period = None

# Exemple 1
D4 = [sites, service, '2018-01-01', '2018-01-31', None, None, None, pcode, None]
```

```python
groundwater = hf.NWIS(D4[0],D4[1],D4[2],D4[3],D4[4],D4[5],D4[6],D4[7],D4[8])
groundwater.get_data().df().mean()
```

```python
# 380616075380701 -  Depth to water level, feet below land surface    1.215141
# 394008077005601 -  Depth to water level, feet below land surface    3.205391
```

```python
# Let's view the data in a specially styled graph!
groundwater.df().plot(marker='o', mfc='white', ms=4, mec='black', color='black')
```

## Download all streamflow data for the state of Virginia

```python
# Use this carefully! You can easily request more data 
# than you will know what to do with.
start = "2017-01-01"
end = "2017-12-31"
vas = hf.get_nwis(None, "dv", start, end, stateCd='va')
```

```python
print(vas.ok)
print(vas.status_code)
print()
my_dict_v = vas.json()
print(my_dict_v)
print()
print(vas.headers)
```

```python
# Extract a dataframe from the response we received from the NWIS.
vad_all = hf.extract_nwis_df(my_dict_v)
vad_all
# vad_all = hf.extract_nwis_df(vas)
```

```python
# Calculate the mean for each site.
vad_all.mean()
```

```python
# There are so many sites that we can't read them all!
# Plot all streamflow data for the state of Virginia
# TROUVER LE NOM DES AXES
vad_all.plot(legend=None, title=None)  # legend=None, == all the wells
```
