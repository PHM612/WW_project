# Modeling
- Statistics using Machine Learning
- Prediction with time series or others

- BOW and TF-IDF are really powerful features. But they do not preserve order of the sequence. 
- For a BOW, the two following sentences are the same:
-    -- "The food is good, not bad at all!"
-    -- "The food is bad, not good at all!"
- RNN :
    
-    -- handle variable length sequences
-    -- track long-term dependencies
-    -- keep information about order
-    -- share parameters across the sequence
- A one to one RNN is a MLP
